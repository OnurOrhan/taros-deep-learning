```
import glob
import pickle
import matplotlib.pyplot as plt
```

```
path = "./history/ctc_large_images/*"
histories = glob.glob(path)
history_omit_1 = []
# print(history_omit_1)
medium_names = []
for history_file in histories[:12]:
  with open(history_file, 'rb') as f:
    c =pickle.load(f)
    history_omit_1.append(c)
  medium_names.append(history_file.split('/')[-1].split('_')[0])
print(medium_names)
```

```
for history in history_omit_1:
  plt.plot(history['loss'])
  plt.title('model training loss')
  plt.ylabel('loss')
  plt.xlabel('epoch')
plt.legend(medium_names, loc='best')
plt.show()
```

```
for history in history_omit_1:
  plt.plot(history['val_loss'])
  plt.title('model validation loss')
  plt.ylabel('loss')
  plt.xlabel('epoch')
plt.legend(medium_names, loc='best')
plt.show()
```

```
file_name = "./history/ctc_large_images/guitar1_guitar3_guitar2_history"
with open(file_name, 'rb') as f:
    history =pickle.load(f)

plt.plot(history['loss'])
plt.plot(history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='best')

```
